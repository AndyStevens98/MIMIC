```python
import pandas as pd
```

```python
map_df = pd.read_csv('../crosswalk_csv/d_labitems_to_loinc.csv')
with open('missing_measurement_mappings.csv') as fin:
    missing_mci = [x.rstrip('\n') for x in fin.readlines()]
format_str = 'ALTER TABLE measurement SET measurement_concept_id = {mci} WHERE measurement_concept_id = 0 AND measurement_source_value = {msv}'

mapping_dict = {}
for code in missing_mci:
    if '|' in code:
        codes = code.split('|')
        try:
            omop_concept_id = list(map_df[map_df['itemid (omop_source_code)']==int(codes[0])].omop_concept_id)[0]
        except IndexError:
            try:
                omop_concept_id = list(map_df[map_df['itemid (omop_source_code)']==int(codes[1])].omop_concept_id)[0]
            except IndexError:
                continue
        if not pd.isna(omop_concept_id):
            mapping_dict[code] = omop_concept_id
    else:
        try:
            omop_concept_id = list(map_df[map_df['itemid (omop_source_code)']==int(code)].omop_concept_id)[0]
        except IndexError:
            continue
        if not pd.isna(omop_concept_id):
            mapping_dict[code] = omop_concept_id
```

```python
full_sql_string = ''
for key, value in mapping_dict.items():
    filled_in_str = format_str.format(mci=value, msv=key)
    full_sql_string += f'\n{filled_in_str}'
full_sql_string
```
